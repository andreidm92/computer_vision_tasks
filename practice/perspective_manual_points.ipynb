## üìã –û–ø–∏—Å–∞–Ω–∏–µ
–í —ç—Ç–æ–º –Ω–æ—É—Ç–±—É–∫–µ –º—ã –≤—Ä—É—á–Ω—É—é –≤—ã–±–µ—Ä–µ–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã 4 —É–≥–ª–æ–≤ –æ–±—ä–µ–∫—Ç–∞ –Ω–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–∏ –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –ø–µ—Ä—Å–ø–µ–∫—Ç–∏–≤–Ω–æ–≥–æ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è.

## üõ† –ü–ª–∞–Ω —Ä–∞–±–æ—Ç—ã
1. –ó–∞–≥—Ä—É–∑–∏—Ç—å –∏ –ø–æ–∫–∞–∑–∞—Ç—å –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ
2. –í–∏–∑—É–∞–ª—å–Ω–æ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å 4 —É–≥–ª–∞
3. –í–≤–µ—Å—Ç–∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –≤—Ä—É—á–Ω—É—é
4. –í—ã–ø–æ–ª–Ω–∏—Ç—å –ø–µ—Ä—Å–ø–µ–∫—Ç–∏–≤–Ω–æ–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ

## üìö –¢–µ–æ—Ä–∏—è
–ü–µ—Ä—Å–ø–µ–∫—Ç–∏–≤–Ω–æ–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –ø–æ–∑–≤–æ–ª—è–µ—Ç "–≤—ã–ø—Ä—è–º–∏—Ç—å" –æ–±—ä–µ–∫—Ç—ã –Ω–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–∏, —Å–Ω—è—Ç—ã–µ –ø–æ–¥ —É–≥–ª–æ–º. –ò—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –º–∞—Ç—Ä–∏—Ü–∞ 3√ó3, –ø–æ—Å—Ç—Ä–æ–µ–Ω–Ω–∞—è –ø–æ 4 –ø–∞—Ä–∞–º —Ç–æ—á–µ–∫.

```python
# –®–∞–≥ 1: –ò–º–ø–æ—Ä—Ç—ã
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# –®–∞–≥ 2: –ó–∞–≥—Ä—É–∑–∫–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
img = cv2.imread('Paper.jpg')  # –ó–∞–º–µ–Ω–∏—Ç—å –Ω–∞ –ø—É—Ç—å –∫ —Å–≤–æ–µ–º—É –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—é
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# –ü–æ–∫–∞–∑ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
plt.figure(figsize=(8,6))
plt.imshow(img)
plt.title("–û–ø—Ä–µ–¥–µ–ª–∏ 4 —É–≥–ª–∞ –æ–±—ä–µ–∫—Ç–∞")
plt.axis('on')
plt.show()
```

```python
# –®–∞–≥ 3: –í–≤–æ–¥ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –≤—Ä—É—á–Ω—É—é
# –ü—Ä–∏–º–µ—Ä –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç (–∑–∞–º–µ–Ω–∏ –Ω–∞ —Å–≤–æ–∏ –ø–æ—Å–ª–µ –ø—Ä–æ—Å–º–æ—Ç—Ä–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è)
points = np.float32([
    [80, 80],   # –õ–µ–≤—ã–π –≤–µ—Ä—Ö–Ω–∏–π —É–≥–æ–ª
    [300, 25],   # –ü—Ä–∞–≤—ã–π –≤–µ—Ä—Ö–Ω–∏–π —É–≥–æ–ª
    [235, 335],   # –õ–µ–≤—ã–π –Ω–∏–∂–Ω–∏–π —É–≥–æ–ª
    [510, 230]    # –ü—Ä–∞–≤—ã–π –Ω–∏–∂–Ω–∏–π —É–≥–æ–ª
])

# –ö–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –ø–æ—Å–ª–µ —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∞—Ü–∏–∏ (—Ä–∞–∑–º–µ—Ä –Ω–æ–≤–æ–≥–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è)
pts2 = np.float32([
    [0, 0],
    [300, 0],
    [0, 400],
    [300, 400]
])
```

```python
# –®–∞–≥ 4: –ü–µ—Ä—Å–ø–µ–∫—Ç–∏–≤–Ω–æ–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ
M = cv2.getPerspectiveTransform(points, pts2)
warped = cv2.warpPerspective(img, M, (300, 400))

# –ü–æ–∫–∞–∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
plt.figure(figsize=(8,6))
plt.imshow(warped)
plt.title("–í—ã–ø—Ä—è–º–ª–µ–Ω–Ω–æ–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ")
plt.axis('off')
plt.show()
```

## ‚úÖ –í—ã–≤–æ–¥—ã
- –ù–∞—É—á–∏–ª—Å—è –≤—Ä—É—á–Ω—É—é –≤—ã–±–∏—Ä–∞—Ç—å –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã —É–≥–ª–æ–≤ –Ω–∞ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–∏.
- –ü—Ä–æ–≤—ë–ª –ø–µ—Ä—Å–ø–µ–∫—Ç–∏–≤–Ω–æ–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —á–µ—Ä–µ–∑ OpenCV.
- –ü–æ–ª—É—á–∏–ª –≤—ã–ø—Ä—è–º–ª–µ–Ω–Ω–æ–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–µ.
