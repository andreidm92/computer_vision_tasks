```python
import cv2
import numpy as np

# 1.  
image = cv2.imread('/content/monkey.jpeg')
if image is None:
    raise ValueError("  .  !")

image.view()
```

```python
(h, w) = image.shape[:2]
```

```python
# 2.     
shift_x = 50
shift_y = 30
angle = 30

# 3.    (  )
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, angle, 1.0)

print(M)
print(M[0, 2])
print(M[1, 2])
#   :
M[0, 2] += shift_x
M[1, 2] += shift_y
```

```python

# 4.   
transformed = cv2.warpAffine(image, M, (w, h))

# 5.  
image.view()
transformed.view()

```
