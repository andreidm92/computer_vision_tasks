```python
import cv2
import numpy as np

# 1. Загрузим изображение
image = cv2.imread('/content/monkey.jpeg')
if image is None:
    raise ValueError("Изображение не загружено. Проверь путь!")

image.view()
```

```python
(h, w) = image.shape[:2]
```

```python
# 2. Зададим параметры сдвига и поворота
shift_x = 50
shift_y = 30
angle = 30

# 3. Получим матрицу поворота (сразу включим сдвиг)
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, angle, 1.0)

print(M)
print(M[0, 2])
print(M[1, 2])
# Добавим сдвиг вручную:
M[0, 2] += shift_x
M[1, 2] += shift_y
```

```python

# 4. Применим аффинное преобразование
transformed = cv2.warpAffine(image, M, (w, h))

# 5. Показать результаты
image.view()
transformed.view()

```
